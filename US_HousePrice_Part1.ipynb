# Preparing Dataset

```python
import numpy as np
import pandas as pd
```

```python
df_CS = pd.read_csv("CSUSHPISA.csv")
df_CS["DATE"] = pd.to_datetime(df_CS["DATE"])
df_CS = df_CS.rename(columns={'CSUSHPISA': 'CS_Index'})
df_CS.reset_index(inplace = True)
df_CS.drop(columns = ["index"], inplace = True)
df_CS
```

```python
df_CS["Year"] = pd.DatetimeIndex(df_CS["DATE"]).year
df_CS["Month"] = pd.DatetimeIndex(df_CS["DATE"]).month
```

```python
df_CS
```

```python
df_unemp = pd.read_csv("UNRATE.csv")
df_unemp
```

```python
# LREM64TTUSM156S

df_emp = pd.read_csv("LREM64TTUSM156S.csv")
df_emp = df_emp.rename(columns={'LREM64TTUSM156S': 'EmpRate'})
df_emp
```

```python

df_GDP = pd.read_csv('GDP (1).csv')
df_GDP
```

```python
df_InterestRate = pd.read_csv('FEDFUNDS.csv')
df_InterestRate = df_InterestRate.rename(columns = {'FEDFUNDS' : 'InterestRate'})
```

```python
df_InterestRate.tail()
```

```python
df_ConsMat = pd.read_csv('WPUSI012011.csv')
df_ConsMat = df_ConsMat.rename(columns = {'WPUSI012011' : 'ConstMaterial'})
df_ConsMat
```

```python
df_CPI = pd.read_csv('CPIAUCSL.csv')
df_CPI = df_CPI.rename(columns = {'CPIAUCSL' : 'CPI'})
df_CPI
```

```python
df_NewMonthlyHouse = pd.read_csv('MSACSR.csv')
df_NewMonthlyHouse = df_NewMonthlyHouse.rename(columns = {'MSACSR' : 'New Monthly House'})
df_NewMonthlyHouse                                                         
                                                          
```

```python
df_subsidy = pd.read_csv('L312051A027NBEA.csv')
df_subsidy = df_subsidy.rename(columns = {'L312051A027NBEA' : 'Subsidy'})
df_subsidy
```

```python
df_WorPop = pd.read_csv('LFWA64TTUSM647S.csv')
df_WorPop = df_WorPop.rename (columns = {'LFWA64TTUSM647S' : 'WorkingPop'})
df_WorPop
```

```python
# Real Median Household Income
df_MedianHouseIncome = pd.read_csv('MEHOINUSA672N.csv')
df_MedianHouseIncome = df_MedianHouseIncome.rename (columns = {'MEHOINUSA672N' : 'Median Household Income'})
df_MedianHouseIncome
```

```python
df_TotalHousehold = pd.read_csv('TTLHH.csv')
df_TotalHousehold = df_TotalHousehold.rename(columns = {'TTLHH' : 'Total_Household'})
df_TotalHousehold
```

```python
# Merging Per Capita GDP (Quarterly data)
df_GDP["DATE"] = pd.to_datetime(df_GDP["DATE"])
df_CS = pd.merge(df_CS,df_GDP, how = "left")
df_CS.head()
```

```python
# Concating dataframes having monthly data to create one dataframe
df = pd.DataFrame()
df_bymonth = [df_CS, df_unemp, df_emp, df_InterestRate, df_ConsMat, df_CPI, df_NewMonthlyHouse, df_WorPop]

for df1 in df_bymonth:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.head()
```

```python
# Merging other dataframes 
others = [ df_subsidy, df_MedianHouseIncome, df_TotalHousehold]

# df_bymonth = [ df_GDP, , df_MedianHouseIncome, ]


for df1 in others:
    if "Year" not in df1.columns:
        df1["Year"] = pd.DatetimeIndex(df1["DATE"]).year
        df1.set_index("DATE", inplace = True)
        df = pd.merge(df, df1, how = "left", on = "Year")
    else:
        df1.set_index("DATE", inplace = True)
        df = pd.merge(df, df1, how = "left", on = "Year")
df["DATE"] = df_CS["DATE"]
df.set_index("DATE", inplace = True)
df.head()
```

```python
df.to_csv("HousePrice_Data.csv")
```
